```python
import cv2
from matplotlib import pyplot as plt

# Make sure the filename is correct
img = cv2.imread("Demon_slayer.webp")  # Replace with actual file name

if img is None:
    print("Image not loaded. Check the file name!")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title("Uploaded Image using OpenCV")
    plt.show()


```
